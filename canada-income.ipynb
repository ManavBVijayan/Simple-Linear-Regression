```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
data = pd.read_csv("canada_income.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
sns.scatterplot(x="year",y="per capita income (US$)",data=data, color="red" )
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Scatter Plot of Per Capita Income Over Time')
plt.show()
```

```python
X = data["year"].values.reshape(-1, 1)
y = data["per capita income (US$)"].values 
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(X,y)
```

```python
y_pred=model.predict(X)
```

```python
sns.scatterplot(x="year",y="per capita income (US$)",data=data, color="red" )
sns.lineplot(x="year",y=y_pred,data=data,color="green")
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Simple Linear Regression')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
mae = mean_absolute_error(y, y_pred)
print(f'Mean Absolute Error (MAE): {mae}')
mse = mean_squared_error(y, y_pred)
print(f'Mean Squared Error (MSE): {mse}')
r2 = r2_score(y, y_pred)
print(f'R-squared: {r2}')
```
